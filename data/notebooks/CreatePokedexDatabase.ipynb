```python
import pandas as pd
import numpy as np
import chardet
import os
import sqlite3 as db
```

```python
csv_files = ["Galar.csv","Hoenn.csv","Johto.csv","Kalos.csv","Kanto.csv","unknown.csv","Sinnoh.csv","Unova.csv","Hisui.csv"]
csv_file = "Alola.csv"

pokedex_dataframe = pd.read_csv(csv_file)


for file_name in csv_files:
  temp_dataframe = pd.read_csv(file_name)
  pokedex_dataframe = pokedex_dataframe.append(temp_dataframe)

pokedex_dataframe1 = pokedex_dataframe.sort_values(['num'], ascending=[True])
display(pokedex_dataframe1)
```

```python
pokedex_dataframe1.reset_index(inplace = True, drop = True)
display(pokedex_dataframe1)
```

```python
pokedex_dataframe1.to_csv("pokedex_dataframe.csv", index=False)
```

```python
def create_database(my_dataframe): 
  conn = db.connect("pokedex_database.db")
  my_dataframe.to_sql('pokedex', conn)
  cursor = conn.cursor()
  data = cursor.execute('''SELECT * FROM pokedex''')
  for column in data.description:
    print(column[0])
  return cursor, conn

def run_database_query(cursor, sql_string):
  my_data = cursor.execute(sql_string)
  cols = [column[0] for column in my_data.description]
  results = pd.DataFrame.from_records(data = my_data.fetchall(), columns = cols)

  return results
```

```python
cursor, conn = create_database(pokedex_dataframe1)
```

```python
alola_results = run_database_query(cursor, "SELECT * FROM pokedex WHERE region='Hoenn'")
print(alola_results)
```
